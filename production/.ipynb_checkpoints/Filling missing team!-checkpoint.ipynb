```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime
import warnings
import requests
import json
import sys
import os
import urllib2
import ast

pd.set_option('display.max_columns', 500)
warnings.filterwarnings('ignore')
```

```python
missing_game_id = "0021700312"
test = "http://data.nba.com/data/v2015/json/mobile_teams/nba/2017/scores/gamedetail/{}_gamedetail.json".format(missing_game_id)
```

```python
file_open = urllib2.urlopen(test).read()
ddata = ast.literal_eval(file_open)

away = pd.DataFrame(ddata['g']['vls']['pstsg'])
home = pd.DataFrame(ddata['g']['hls']['pstsg'])
```

```python
away.columns = [i.upper() for i in away.columns]
away['PLAYER_NAME'] = away['FN'] + " " + away['LN']
away.rename(columns={"POS":"START_POSITION", 'PID': "PLAYER_ID", "PM":'PLUS_MINUS', 'TOV':"TO",
                    'TPA':'FG3A', 'TPM':"FG3M", "MEMO":"COMMENT"}, inplace=True)
away["GAME_ID"] = ddata['g']['gid']
away["TEAM_ABBREVIATION"] = ddata['g']['vls']['ta']
away["TEAM_ID"] = ddata['g']['vls']['tid']
away["TEAM_NAME"] = ddata['g']['vls']['tn']
away["TEAM_CITY"] = ddata['g']['vls']['tc']
away["FG_PCT"] = away['FGA'] / away['FGA']
away["FG3_PCT"] = away['FG3M'] / away['FG3A']
away["FT_PCT"] = away['FTM'] / away['FTA']

for i in ["FG_PCT", "FG3_PCT", "FT_PCT"]:
    away[i] = away[i].fillna(0)

home.columns = [i.upper() for i in home.columns]
home['PLAYER_NAME'] = home['FN'] + " " + home['LN']
home.rename(columns={"POS":"START_POSITION", 'PID': "PLAYER_ID", "PM":'PLUS_MINUS', 'TOV':"TO",
                    'TPA':'FG3A', 'TPM':"FG3M", "MEMO":"COMMENT"}, inplace=True)
home["GAME_ID"] = ddata['g']['gid']
home["TEAM_ABBREVIATION"] = ddata['g']['hls']['ta']
home["TEAM_ID"] = ddata['g']['hls']['tid']
home["TEAM_CITY"] = ddata['g']['hls']['tc']
home["TEAM_NAME"] = ddata['g']['hls']['tn']
home["FG_PCT"] = home['FGA'] / home['FGA']
home["FG3_PCT"] = home['FG3M'] / home['FG3A']
home["FT_PCT"] = home['FTM'] / home['FTA']

for i in ["FG_PCT", "FG3_PCT", "FT_PCT"]:
    home[i] = home[i].fillna(0)
```

### MAIN_TEAM TABLE 

```python
main_team = pd.read_csv("main_players_2017.csv")
```

```python
new = away.append(home)
new_main_team = main_team.append(new)
new_main_team = new_main_team[main_team.columns]
```
